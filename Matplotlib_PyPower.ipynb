## Introduction

#### - Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy.
#### - Pyplot is a Matplotlib module which provides a MATLAB-like interface.
#### - Matplotlib is designed to be as usable as MATLAB, with the ability to use Python, and the advantage of being free and open-source.
#### - make it easy to produce static raster or vector graphics files without the need for any GUIs.

## Toolkit

####         Several toolkits are available which extend Matplotlib functionality. Some are separate downloads, others ship with the Matplotlib source code but have external dependencies.

#### - Basemap: map plotting with various map projections, coastlines, and political boundaries
#### - Cartopy: a mapping library featuring object-oriented map projection definitions, and arbitrary point, line, polygon and image transformation capabilities.(Matplotlib v1.2 and above)
#### - Excel tools: utilities for exchanging data with Microsoft Excel
#### - GTK tools: interface to the GTK+ library
#### - Qt interface
#### - Mplot3d: 3-D plots
#### - Natgrid: interface to the natgrid library for gridding irregularly spaced data.
#### - matplotlib2tikz: export to Pgfplots for smooth integration into LaTeX documents

## 2)Axes

All plotting is done with respect to an Axes. An Axes object must belong to a Figure (and only one Figure). Most commands you will ever issue will be with respect to this Axes object.

Typically, you'll set up a Figure, and then add an Axes to it.

You can use fig.add_axes, but in most cases, you'll find that adding a subplot will fit your needs perfectly. (Again a "subplot" is just an axes on a grid system.)

```
fig = plt.figure()
```

The purpose of using **plt.figure()** is to create a figure object.

You can assume a figure object as a frame.

The whole figure is regarded as the figure object. It is necessary to explicitly use **plt.figure()** when we want to tweak the size of the figure and when we want to add multiple Axes objects in a single figure.

```
axe = fig.add_subplot(111) 
```

**add_subplot(111)** are subplot grid parameters encoded as a single integer. For example, "111" means "1x1 grid, first subplot" and "234" means "2x3 grid, 4th subplot".

Alternative form for **add_subplot(111)** is **add_subplot(1, 1, 1)**

### Plotting Axes :

#### basic : 

First let's print the x-y axis.

```
fig = plt.figure()
# setting the size of canvas
axes = fig.add_axes([1,1,1,1])
 
# plotting the graph
axes.plot()
 
# displaying the graph
plt.show()
```

To plot the points on axis :
here **plt.scatter(x,y)** will plot the points ,We will see about scatter ,soon.

And to show the graph we will use **plt.show()** function.

```
fig = plt.figure()
axes = fig.add_axes([1,1,1,1])
# plotting the graph
x=[1,2,3,4,5,6]
y=[2,3,4,5,6,7]
axes.scatter(x,y)

x=[1,2,3,4,5,6]
y=[1,2,3,4,5,6]
axes.scatter(x,y)
 
# displaying the graph
plt.show()
```

To make a graph with steps we will use **axes.step()**

```
fig = plt.figure()
axes = fig.add_axes([1,1,1,1])
x=[1,2,3,4,5,6]
y=[2,4,9,16,25,36]
# plotting the graph
axes.step(x,y,'g-*')
 
# displaying the graph
plt.show()
```

#### Spans :

To plot the span on axes we have 4 functions :

-Axes.axhline	Add a horizontal line across the axis.
 

-Axes.axhspan	Add a horizontal span (rectangle) across the axis.


-Axes.axvline	Add a vertical line across the axes.


-Axes.axvspan	Add a vertical span (rectangle) across the axes.

```
fig = plt.figure()
axes = fig.add_axes([1,1,1,1])
 
# plotting the graph
axes.axhline(25) #marks line at y=25
plt.show()
```

```
fig = plt.figure()
axes = fig.add_axes([1,1,1,1])
axes.axvspan(0,30) #Takes min y and max y as attributes
plt.show()
```

```
fig = plt.figure()
axes = fig.add_axes([1,1,1,1])
axes.axvline(10) #marks line at x=10
plt.show()
```

```
fig = plt.figure()
axes = fig.add_axes([1,1,1,1])
axes.axhspan(0,50) #Takes min x and max x as attributes
plt.show()
```

```
fig = plt.figure(figsize=(10,4))
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.scatter(x, y)
plt.show()
```

To resize the figure use the parameter **figsize=(x,y)** ,here x and y are heigth and width

There are many more advanced function of matplotlib axes ,that you can see [here](https://matplotlib.org/api/axes_api.html).

## 3)Ploting Graphs
**Line :** Now Let's plot a line *x=y* in a graph.

```
plt.plot([1,2,3,4,5,6,7,8,9],[1,2,3,4,5,6,7,8,9]) #just make a list of x and y points to be plotted
plt.show()
```

Great Isn't it?

You can change the x and y points !

```
plt.plot([3,5,7],[5.8,9.9,-0.5])
plt.show()
```

In matplotlib you can put labels to remember what graph is stands for.

Lables can be used to name x and y axes and also different graphs and also you can give **titles** to the graph by **plt.title()** function.

```
plt.title('Demo graph\nTwo Lines')
plt.xlabel('X axes')
plt.ylabel('Y axes')
plt.plot([3,5,7],[5.8,9.9,-0.5],label='First line')
plt.plot([1,2,3],[2,3,6],label='Second line')
plt.show()
```

Here we can see the title and the axes names but there is no label of two lines shown in above graph.
Because we haven't use the **legend()** function.
**plt.legend()** : Matplotlib has native support for legends. Legends can be placed in various positions: A legend can be placed inside or outside the chart and the position can be moved.

The legend() method adds the legend to the plot.
Here the ***box*** containing labels of two lines of graphs is showing the legend.

```
plt.title('Demo graph\nTwo Lines')
plt.xlabel('X axes')
plt.ylabel('Y axes')
plt.plot([3,5,7],[5.8,9.9,-0.5],label='First line')
plt.plot([1,2,3],[2,3,6],label='Second line')
plt.legend()
plt.show()
```

You can put this legend inside or outside a graph and you can also edit the legend.
For example i put the legend outside the graph :

```
plt.title('Demo graph\nTwo Lines')
plt.xlabel('X axes')
plt.ylabel('Y axes')
plt.plot([3,5,7],[5.8,9.9,-0.5],label='First line')
plt.plot([1,2,3],[2,3,6],label='Second line')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), shadow=True, ncol=2)
plt.show()
```

here **ncol=2** shows number of coloums in the legend.
**bbox_to_anchor=(0.5, -0.05)** is showing the coordinates of the legend to put in.
### Graph Colors : 
You can give different colors to your graphs by jus passing **color='First letter of any color'**.

Let's take any graph as example and chenge it's colors.


```
plt.title('Demo graph\nTwo Lines')
plt.xlabel('X axes')
plt.ylabel('Y axes')
plt.plot([3,5,7],[5.8,9.9,-0.5],label='First line', color='b') #black color
plt.plot([1,2,3],[2,3,6],label='Second line' , color='r') #red color
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), shadow=True, ncol=2)
plt.show()
```

## 4)Different types of Graphs in matplotlib
**Bar chart :** for ploting in bar charts we have to use **plt.bar()** function.

```
x=[1,2,3,4,5]
y=[2,4,9,16,25]
plt.title('Bar Graphs')
plt.bar(x,y,label='Demo Bar Chart')
plt.legend()
plt.show()
```

You can definately plot two or more bar graphs in this.
And two make diffrence beetween them you can change the colours of the graphs.
to change color we only have to use only one letter of that color's name ,For example
**Blue** --> 'B'
**Red** --> 'R'
**Black** --> 'K'
**Cyan** --> 'C'

```
x=[1,2,3,4,5]
y=[2,4,9,16,25]
x2=[1,2,3,4,5]
y2=[1,2,3,4,5]
plt.title('Bar Graphs')
plt.bar(x,y,label='Bar 1',color='c')
plt.bar(x2,y2,label='Bar 2',color='b')
plt.legend()
plt.show()
```

**Histograms** : To generate a 1D histogram we only need a single vector of numbers.

For a 2D histogram we'll need a second vector. We'll generate both below, and show the histogram for each vector.

```
# Y-axis values 
y = [12, 3,6 ,1,9,5,4,4] 
  
# Function to plot histogram
plt.hist(y,color='R') 
  
# function to show the plot 
plt.show() 
```

**Stack Plots :** Stack graph is just what you are thinking of "Stacks".You can plot diiferent values of any variable to plot a stack for better visualization by **plt.stackplot()** function.

Let's say i made a weekly report on how many hours I spent on the activities :

```
days=[1,2,3,4,5,6,7]
sleeping=[7,8,6,11,8,9,10]
eating=[3,2,1,2,3,1,2]
learning_from_PYPOWER=[8,10,5,2,0,10,9]
exersicing=[1,2,1,0,0,1,0]

plt.stackplot(days,sleeping,eating,learning_from_PYPOWER,exersicing,colors=['R','C','B','K'])
plt.title('Weekly Report')
plt.xlabel('Activities')
plt.ylabel('Spent Hours')
plt.legend()
plt.show()
```

Ohh! I am sleeping a lot nowadays. I have to do something.

**Pie charts** : To use Pie charts we can use **plt.pie()** function.
Let's plot above graph in to pie charts.

```
slice=[1,2,3,4]
sleeping=[7,8,6,11,8,9,10]
eating=[3,2,1,2,3,1,2]
learning_from_PYPOWER=[8,10,5,2,0,10,9]
exersicing=[1,2,1,0,0,1,0]

plt.pie(slice,labels=['Sleeping','Eating','learning_from_PYPOWER','exersicing'],colors=['R','C','B','K'])
plt.title('Weekly Report')
plt.xlabel('Activities')
plt.ylabel('Spent Hours')
plt.legend(loc='upper center', bbox_to_anchor=(0.1,0), shadow=True, ncol=2)
plt.show()
```

Now,if you want to edit the pie chart you can do by **shadow** function used above.
And to take piece from pie chart you can use **explode** function.For example **explode(0,0,0.1,0)** this will take third chart "learning_from_PYPOWER" 0.1  spaced outside.

```
slice=[1,2,3,4]
sleeping=[7,8,6,11,8,9,10]
eating=[3,2,1,2,3,1,2]
learning_from_PYPOWER=[8,10,5,2,0,10,9]
exersicing=[1,2,1,0,0,1,0]

plt.pie(slice,labels=['Sleeping','Eating','learning_from_PYPOWER','exersicing'],colors=['R','C','B','K'], shadow=True , explode=(0,0,0.1,0))
plt.title('Weekly Report')
plt.xlabel('Activities')
plt.ylabel('Spent Hours')
plt.legend(loc='upper center', bbox_to_anchor=(0.1,0), shadow=True, ncol=2)
plt.show()
```
